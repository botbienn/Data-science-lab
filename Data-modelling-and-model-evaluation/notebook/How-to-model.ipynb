# 1. Các bước thực hiện

## 1.1 Chọn mô hình phù hợp:

Học có giám sát (Supervised learning):

- Hồi quy (Regression): Sử dụng cho các bài toán dự đoán giá trị liên tục. Ví dụ: linear regression, logistic regression.

- Phân loại (Classification): Sử dụng cho các bài toán phân loại dữ liệu vào các nhóm cụ thể. Ví dụ: decision tree, random forest, SVM, neural network.

Học không giám sát (Unsupervised learning):

- Phân cụm (Clustering): Sử dụng để nhóm các điểm dữ liệu thành các cụm dựa trên sự tương đồng. Ví dụ: K-means, DBSCAN, Hierarchical clustering.

- Giảm chiều (Dimensionality Reduction): Sử dụng để giảm số lượng biến trong dữ liệu. Ví dụ: PCA, t-SNE.

## 1.2 Chuẩn bị dữ liệu:

Chia dữ liệu: Chia dữ liệu thành tập huấn luyện (training set) và tập kiểm tra (testing set) để đánh giá hiệu quả của mô hình.

Xử lý dữ liệu thiếu: Xử lý các giá trị bị thiếu bằng cách loại bỏ hoặc thay thế chúng.

Chuẩn hóa dữ liệu: Biến đổi dữ liệu để đảm bảo các biến có cùng thang đo hoặc phân phối.

## 1.3 Huấn luyện mô hình (Training the model):

Sử dụng tập dữ liệu huấn luyện để huấn luyện mô hình. Quá trình này liên quan đến việc tối ưu hóa các tham số của mô hình để giảm thiểu sai số dự đoán.

## 1.4 Đánh giá mô hình (Model Evaluation):

Đánh giá hiệu suất: Sử dụng các chỉ số như độ chính xác (accuracy), độ nhạy (recall), độ đặc hiệu (precision), F1 score, và MSE (Mean Squared Error) để đánh giá hiệu suất của mô hình trên tập dữ liệu kiểm tra.

Kỹ thuật kiểm định chéo (Cross-validation): Sử dụng để đánh giá mô hình một cách toàn diện hơn bằng cách chia dữ liệu thành nhiều tập con và kiểm tra mô hình trên mỗi tập con.

## 1.5 Tinh chỉnh và tối ưu hóa mô hình (Model Tuning and Optimization):

Tinh chỉnh tham số (Hyperparameter Tuning): Tối ưu hóa các tham số của mô hình để đạt được hiệu suất tốt nhất. Có thể sử dụng các phương pháp như Grid Search hoặc Random Search.

Chọn mô hình cuối cùng (Model Selection): Chọn mô hình có hiệu suất tốt nhất dựa trên các đánh giá và kiểm tra.

## 1.6 Triển khai mô hình (Model Deployment):

Sau khi chọn được mô hình tốt nhất, tiến hành triển khai mô hình vào hệ thống thực tế để sử dụng cho dự đoán và phân tích dữ liệu.

## 1.7 Theo dõi và bảo trì mô hình (Model Monitoring and Maintenance):

Theo dõi hiệu suất của mô hình sau khi triển khai và thực hiện các cập nhật cần thiết để đảm bảo mô hình vẫn hoạt động hiệu quả với dữ liệu mới.

# 2. Chia dữ liệu

## 2.1 Tập huấn luyện (Training Set)

Định nghĩa: Tập huấn luyện là phần dữ liệu được sử dụng để huấn luyện mô hình.

Mục đích: Mô hình học từ tập huấn luyện để tìm ra các mẫu hình và mối quan hệ trong dữ liệu.

Chi tiết: Tập huấn luyện thường chiếm phần lớn dữ liệu ban đầu, thường từ 60% đến 80% của toàn bộ dữ liệu.

## 2.2 Tập kiểm tra (Test Set)

Định nghĩa: Tập kiểm tra là phần dữ liệu được giữ lại để đánh giá hiệu suất của mô hình sau khi đã huấn luyện.

Mục đích: Đánh giá mô hình để đảm bảo rằng nó không chỉ hoạt động tốt trên tập huấn luyện mà còn trên dữ liệu mới, chưa từng thấy trước đó.

Chi tiết: Tập kiểm tra thường chiếm từ 20% đến 40% của toàn bộ dữ liệu ban đầu.

## 2.3 Tập xác nhận (Validation Set)

Định nghĩa: Tập xác nhận là một phần dữ liệu được tách ra từ tập huấn luyện để điều chỉnh các tham số của mô hình và lựa chọn mô hình tốt nhất.

Mục đích: Đánh giá mô hình trong quá trình huấn luyện và tinh chỉnh các tham số mà không ảnh hưởng đến tập kiểm tra.

Chi tiết: Tập xác nhận thường chiếm từ 10% đến 20% của tập huấn luyện.

## 2.4 Quy trình chia dữ liệu

Chia dữ liệu ban đầu: Tách dữ liệu ban đầu thành tập huấn luyện và tập kiểm tra.

Chia tập huấn luyện: Tách một phần của tập huấn luyện thành tập xác nhận (nếu cần).

## 2.5 Kỹ thuật kiểm định chéo (Cross-Validation)

Định nghĩa: Kỹ thuật kiểm định chéo là một phương pháp để đánh giá mô hình một cách toàn diện hơn bằng cách chia dữ liệu thành nhiều phần và sử dụng mỗi phần một lần làm tập kiểm tra và các phần còn lại làm tập huấn luyện.

Ví dụ phổ biến: K-Fold Cross-Validation là phương pháp chia dữ liệu thành K phần (folds) và lặp lại K lần, mỗi lần sử dụng một fold khác nhau làm tập kiểm tra và các fold còn lại làm tập huấn luyện.

## 2.6 Tại sao cần chia dữ liệu?

Tránh overfitting: Nếu chỉ sử dụng tập huấn luyện, mô hình có thể học quá kỹ các chi tiết và nhiễu trong dữ liệu huấn luyện, dẫn đến hiệu suất kém trên dữ liệu mới.

Đánh giá khách quan: Sử dụng tập kiểm tra giúp đánh giá mô hình một cách khách quan, vì nó không tham gia vào quá trình huấn luyện và tối ưu hóa.

Tinh chỉnh mô hình: Tập xác nhận giúp lựa chọn mô hình và tham số tốt nhất mà không cần sử dụng tập kiểm tra, giúp đảm bảo rằng tập kiểm tra chỉ được sử dụng cho đánh giá cuối cùng.

## 3.1 Dùng thư viện scikit-learn

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import joblib
```

Giả sử đã khám phá, tiền xử lý dữ liệu,... xong trước đó.

Bây giờ mình sẽ xây dựng mô hình để dự đoán nhiệt độ trung bình trong ngày ở TP HCM.

```python
df = pd.read_csv('../data/HoChiMinh.csv')
df.head()
```

```python
# bỏ các thuộc tính không cần khi train
df = df.drop(columns=['Address', 'Datetime', 'DatetimeEpoch', 'Preciptype'])
df.columns
```

```python
# Thuộc tính mình cần dự đoán
target = 'Temp'
# Lấy tên các thuộc tính
features = df.drop(columns=target).columns

X = df[features]
Y = df[target]

# 20% dữ liệu làm Test set
# seed random = 22120182
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=22120182)

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X_train_normalized, Y_train)

# Dự đoán trên tập kiểm tra
Y_pred = model.predict(X_test_normalized)
```

```python
# Đánh giá mô hình
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f'Mean Squared Error: {mse:.5f}')
print(f'R^2 Score: {r2:.5f}')
print(Y_pred[:5])
```

## 3.2 Không dùng scikit-learn

Tham khảo: https://machinelearningcoban.com/2016/12/28/linearregression/

```python
class LR:
    def __init__(self):
        self.w = None

    def fit(self, X, Y):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        self.w = np.linalg.pinv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)

    def predict(self, X):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]
        return X_b.dot(self.w)

    def mean_squared_error(self, Y_test, Y_pred):
        return ((Y_test - Y_pred) ** 2).mean()

    def r2_score(self, Y_test, Y_pred):
        return 1 - ((Y_test - Y_pred) ** 2).sum() / ((Y_test - Y_test.mean()) ** 2).sum()
```

Để cho tiện trong việc so sánh thì mình dùng lại $X$, $Y$ ở trên luôn :p, còn khi code mọi người có thể tự viết hàm chia training set, testing set với hàm chuẩn hóa.

```python
model = LR()
model.fit(X_train_normalized, Y_train)
Y_pred = model.predict(X_test_normalized)
```

```python
mse = model.mean_squared_error(Y_test, Y_pred)
r2 = model.r2_score(Y_test, Y_pred)

print(f'Mean Squared Error: {mse:.5f}')
print(f'R^2 Score: {r2:.5f}')
print(Y_pred[:5])
```

## 3.3 Đánh giá mô hình hồi quy (Evaluation)

**[MSE (Mean Squared Error)](https://scikit-learn.org/dev/modules/generated/sklearn.metrics.mean_squared_error.html)**: đo lường trung bình của bình phương sai số giữa giá trị thực và giá trị dự đoán. MSE là một thước đo để xác định mức độ dự đoán của mô hình sai lệch so với giá trị thực tế.

\begin{equation*}
MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2
\end{equation*}

**[R^2](https://scikit-learn.org/dev/modules/generated/sklearn.metrics.r2_score.html)**: là một thước đo thống kê phản ánh tỷ lệ phương sai của biến đích được giải thích bởi các biến dự đoán trong mô hình hồi quy. Nó cung cấp một cách để xác định mức độ phù hợp của mô hình.

\begin{equation*}
R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}
\end{equation*}

Trong đó: 

- $n$: số mẫu.

- $y_i$: giá trị thực tế.

- $\hat{y}_i$: giá trị dự đoán.

- $\bar{y}$: giá trị trung bình.
