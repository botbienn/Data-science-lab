```python
import pandas as pd
import numpy as np
import json
from pprint import pprint
```

```python
city = ['LamDong', 'DaNang']
cols_to_save = ['Address', 'Datetime', 'DatetimeEpoch', 'Tempmax', 'Tempmin', 'Temp', 'Dew', 'Humidity', 
                'Precip', 'Precipprob', 'Precipcover', 'Preciptype', 'Windgust', 'Windspeed', 'Winddir', 
                'Pressure', 'Cloudcover', 'Visibility', 'Solarradiation', 'Solarenergy', 'Uvindex', 'Moonphase']

def process_city_data(city_name):
    # Hàm xử lý và lưu file CSV từ dữ liệu JSON
    for suffix in ['', '1']:  # Duyệt qua cả hai file: city_name.json và city_name1.json
        # Đọc file JSON với mã hóa utf-8
        with open(f'../DN-LD/{city_name}{suffix}.json', 'r', encoding='utf-8') as file:
            data = json.load(file)

        # Chuyển đổi dữ liệu 'days' thành DataFrame
        df = pd.json_normalize(data['days'])
        df.columns = [col[0].upper() + col[1:] for col in df.columns]  # Viết hoa chữ cái đầu tiên của các cột
        df['Address'] = data['address']  # Thêm cột Address vào DataFrame

        # Ghi vào file CSV (mode='a' để append dữ liệu, header=False để không ghi lại header lần thứ hai)
        df[cols_to_save].to_csv(f'{city_name}.csv', mode='a', index=False, header=(suffix == ''))

# Xử lý dữ liệu cho tất cả các thành phố trong danh sách
for city_name in city:
    process_city_data(city_name)
```
