# Phân tích và Dự đoán Thời tiết

## Câu hỏi 1: Dự đoán tốc độ gió (Windspeed) dựa trên các điều kiện thời tiết?

### Ý nghĩa của câu hỏi:
- Tốc độ gió ảnh hưởng trực tiếp đến an toàn giao thông, hàng không, hàng hải, và các dự án năng lượng gió.
- Dự đoán tốc độ gió giúp lập kế hoạch phòng tránh rủi ro từ thời tiết bất lợi.

### Cách trả lời câu hỏi:
1. **Chọn cột:**
   - **Cột mục tiêu:** `Windspeed`.
   - **Cột đặc trưng:** `Pressure`, `Temp`, `Humidity`, `Cloudcover`, `Visibility`, `Precip`, `Winddir`, `Solarradiation`.
   - **Lý do chọn cột:** Các yếu tố này có ảnh hưởng lớn đến luồng không khí và tốc độ gió.

2. **Kỹ thuật:**
   - **Chuẩn bị dữ liệu:**
     - Xử lý giá trị bị thiếu (missing values) và chuẩn hóa dữ liệu (normalization).
   - **Chọn mô hình học máy:**
     - Các mô hình hồi quy như Random Forest Regression, Gradient Boosting, Neural Networks.
   - **Quy trình:**
     - Chia dữ liệu thành tập huấn luyện (train), kiểm tra (test), và xác thực (validation).
     - Huấn luyện mô hình và đánh giá bằng các chỉ số như MSE (Mean Squared Error), MAE (Mean Absolute Error).

## Câu hỏi 2: Các yếu tố nào là nguyên nhân chính dẫn đến các hiện tượng thời tiết cực đoan?

### Ý nghĩa của câu hỏi:
- Hiểu rõ các yếu tố dẫn đến thời tiết cực đoan (mưa lớn, gió mạnh, nhiệt độ cao) giúp xây dựng hệ thống cảnh báo sớm.
- Hỗ trợ nghiên cứu các biện pháp giảm thiểu tác động của hiện tượng khí hậu cực đoan.

### Cách trả lời câu hỏi:
1. **Xác định hiện tượng cực đoan:**
   - **Mưa lớn:** `Precip` > 90th percentile.
   - **Nhiệt độ cao:** `Tempmax` > 90th percentile.
   - **Gió mạnh:** `Windgust` > 90th percentile.

2. **Chọn cột:**
   - **Biến mục tiêu:** Các hiện tượng cực đoan như `ExtremeRain`, `ExtremeHeat`, `ExtremeWind`.
   - **Biến đặc trưng:** `Temp`, `Dew`, `Humidity`, `Pressure`, `Cloudcover`, `Windspeed`, `Solarenergy`.

3. **Phân tích dữ liệu:**
   - Áp dụng phân tích tương quan để tìm mối liên hệ giữa các yếu tố và hiện tượng cực đoan.
   - Dùng PCA (Principal Component Analysis) để giảm chiều dữ liệu.

4. **Mô hình hóa:**
   - Sử dụng các mô hình phân loại như Random Forest, XGBoost để xác định yếu tố quan trọng nhất.
   - Phân tích Feature Importance để đưa ra kết luận.

## Câu hỏi 3: Xu hướng nhiệt độ trung bình năm của từng thành phố như thế nào? Thành phố nào có tốc độ tăng nhiệt độ nhanh nhất?

### Ý nghĩa của câu hỏi:
- Theo dõi biến đổi khí hậu ở cấp độ địa phương, hỗ trợ chính sách giảm thiểu tác động của biến đổi khí hậu.
- Xác định các khu vực chịu ảnh hưởng nhiều nhất để ưu tiên nguồn lực.

### Cách trả lời câu hỏi:
1. **Chuẩn bị dữ liệu:**
   - Tính nhiệt độ trung bình năm (`AnnualAvgTemp`) cho mỗi thành phố.
   - Áp dụng hồi quy tuyến tính để tính tốc độ tăng nhiệt độ (`slope`) theo năm.

2. **Phân tích xu hướng:**
   - Vẽ biểu đồ chuỗi thời gian của nhiệt độ trung bình từng năm cho từng thành phố.
   - Sử dụng kiểm định Mann-Kendall hoặc Sen's Slope để kiểm tra xu hướng tăng/giảm.

3. **So sánh tốc độ tăng nhiệt độ:**
   - Tìm thành phố có tốc độ tăng nhiệt độ nhanh nhất bằng cách so sánh `slope` giữa các thành phố.

4. **Trực quan hóa:**
   - Sử dụng bản đồ nhiệt (heatmap) để hiển thị tốc độ tăng nhiệt độ theo không gian.
   - Biểu đồ cột để so sánh tốc độ tăng nhiệt độ giữa các thành phố.
